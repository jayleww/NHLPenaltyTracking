```python
import pandas as pd
import os
```

```python
direct = "NHLPenaltyTracking"
```

```python
os.chdir(direct)
```

```python
files = [file for file in os.listdir() if (file.startswith('NHL')) and file.endswith('.png')==False]
```

```python
for file in files:
    f = pd.read_csv(file)
    f.drop('Year', axis=1, inplace=True)
    year = [int(n) for n in file.split('_') if n.isdigit()]
    f['Season'] = year[0]
    f.to_csv(file, header=True)
```
